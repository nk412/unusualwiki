```python
URL = "https://en.wikipedia.org/wiki/Wikipedia:Unusual_articles"
OUTPUT = "./data/wiki.json"
```

```python
import requests
import datetime
import json
from bs4 import BeautifulSoup
```

```python
a = requests.get(URL)
```

```python
b = BeautifulSoup(a.text)
```

```python
tables = b.find_all("table", { "class" : "wikitable" })
results = []
for table in tables:
    for row in table.findAll("tr"):
        cells = row.findAll("td")
        title = cells[0].contents[0]
        
        if not (str(title).startswith('<b><a href=') or str(title).startswith('<td><b><a href=')):
            continue
        
        try:
            desc = cells[1]
        except:
            print(row)
            continue

        title_clean = title.get_text()
        desc_clean = desc.get_text().strip('\n')
        url = title.find('a').get('href').replace('/wiki/', 'https://en.wikipedia.org/wiki/')
        results.append({
            'title': title_clean,
            'desc': desc_clean,
            'url': url
        })

```

```python
with open(OUTPUT, 'w') as f:
    f.write(json.dumps({
        'scraped': datetime.datetime.now().isoformat(),
        'articles': results
    }
    ,sort_keys=True))
```
